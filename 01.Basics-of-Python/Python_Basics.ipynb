# 2. Data types in Python
- python has various data types like integer, float, string, boolean, complex, list, tuple, set, dictionary etc.
- integer data type is used to store whole numbers.
- float data type is used to store decimal numbers.
- string data type is used to store sequence of characters.
- boolean data type is used to store True or False values.
- complex data type is used to store complex numbers.
- list data type is used to store multiple values in a single variable. it is ordered and mutable.
- tuple data type is used to store multiple values in a single variable. it is ordered and immutable.
- set data type is used to store multiple values in a single variable. it is unordered and unindexed.
- dictionary data type is used to store key-value pairs. it is unordered and mutable.
- to check the data type of variable we use type() function.


```python
# python has various data types like integer, float, string, boolean, complex, list, tuple, set, dictionary etc.
# integer data type is used to store whole numbers.
# float data type is used to store decimal numbers.
# string data type is used to store sequence of characters.
# boolean data type is used to store True or False values.
# complex data type is used to store complex numbers.
# list data type is used to store multiple values in a single variable. it is ordered and mutable.
# tuple data type is used to store multiple values in a single variable. it is ordered and immutable.
# set data type is used to store multiple values in a single variable. it is unordered and unindexed.
# dictionary data type is used to store key-value pairs. it is unordered and mutable.
# to check the data type of variable we use type() function.
# example of different data types in python :

x = 10 # integer data type
x = 17.e309 # this will be infinity as it exceeds the float limit
y = 10.5 # float data type
name = "gagan" # string data type
is_student = True # boolean data type
complex_num = 3 + 5j # complex data type
print(x,y,name,is_student,complex_num)
print(type(x),type(y),type(name),type(is_student),type(complex_num))
print([1,2,3,4,5]) # list data type
print((1,2,3,4,5)) # tuple data type
print({1,2,3,4,5}) # set data type
print({"name":"gagan","age":23}) # dictionary data type

a=b=c=10 # multiple variable assignment
print(a,b,c)
a,b,c=10,20,30 # multiple variable assignment with different values
print(a,b,c)
```

## 3. Dynamic Typing in Python

```python
# in c/c++/java we need to declare the data type of variable explicitly. like => int x = 10;
# in python we don't need to declare the data type of variable explicitly. it is dynamically typed language.
# this means that the data type of variable is determined at runtime based on the value assigned to it.
# example of dynamic typing in python:
x = 10 # x is integer
print(x,type(x))
x = "gagan" # now x is string
print(x,type(x))

x = 10.5 # now x is floated
print(x,type(x))
# as we can see the data type of variable x is changing as we are assigning different values to it.
# this is dynamic typing in python.
```

## 4. Dynamic Binding in Python


```python
# in python memory allocation is done at runtime. this is called dynamic binding.
# when we assign a value to a variable, memory is allocated to that variable at runtime.
x = 10 # memory allocated for integer 10
print(x,id(x))
x = "gagan"
print(x,id(x))

x = 10.5
print(x,id(x))
# as we can see the memory address of variable x is changing as we are assigning different values to it.
# this is dynamic binding in python.
```

## 6. Python Type Casting


```python
# type casting is used to convert one data type to another data type in python.
# to convert data type we use built-in functions like int(), float(), str(), bool()
# example of type casting in python:
x = "10" # x is string
print(x,type(x))
x = int(x) # converting string to integer
print(x,type(x))

y = 10.9 # y is floating point number
print(y,type(y))
y = int(y) # converting float to integer
print(y,type(y))

z = 10 # z is integer
print(z,type(z))
z = float(z) # converting integer to float
print(z,type(z))

a = 10 # a is integer
print(a,type(a))
a = str(a) # converting integer to string
print(a,type(a))

b = "True" # b is string
print(b,type(b))
b = bool(b) # converting string to boolean
print(b,type(b))

# more examples of type casting in python:
a = 7
a = int(a)
print(a,type(a))
a = float(a)
print(a,type(a))
a = str(a)
print(a,type(a))
a = bool(a)
print(a,type(a))

b = 1.0
b = int(b)
print(b,type(b))
b = float(b)
print(b,type(b))
b = str(b)
print(b,type(b))
b = bool(b)
print(b,type(b))

# c = "123xyz"
# c = int(c) # this will give error as string contains non-numeric characters
print(c,type(c))
c = "123"
c = int(c)
print(c,type(c))
c = float(c)
print(c,type(c))
c = str(c)
print(c,type(c))
c = bool(c)
print(c,type(c))

d = False
d = int(d)
print(d,type(d))
d = float(d)
print(d,type(d))
d = str(d)
print(d,type(d))
d = bool(d)
print(d,type(d))

# as we can see the data type of variable is changing as we are type casting it to different data types.
```

## 8. Python literals

```python
# literals are the data values that are assigned to variables in python.
# there are different types of literals in python like string literals, numeric literals, boolean literals,
# special literals, collection literals etc.
# string literals are enclosed in single quotes(' ') or double quotes(" ").
str1 = 'hello'
str2 = "gagan"
print(str1,str2)
# numeric literals are of three types: integer literals, float literals, complex literals.
int_literal = 10
float_literal = 10.5
complex_literal = 3 + 5j
print(int_literal,float_literal,complex_literal)
# boolean literals are either True or False.
bool_literal1 = True
bool_literal2 = False
print(bool_literal1,bool_literal2)
# special literals is None which represents the absence of value.
special_literal = None
print(special_literal)
# collection literals are of three types: list literals, tuple literals, set literals, dictionary literals
list_literal = [1,2,3,4,5]
tuple_literal = (1,2,3,4,5)
set_literal = {1,2,3,4,5}
dict_literal = {"name":"gagan","age":23}
print(list_literal)
print(tuple_literal)
print(set_literal)
print(dict_literal)
# as we can see different types of literals in python.
```
